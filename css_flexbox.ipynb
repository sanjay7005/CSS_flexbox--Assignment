Q1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?

CSS Flexbox and Grid are both layout models that help design web pages efficiently, but they serve different purposes and excel in different scenarios.

# **Main Differences: Flexbox**

- One-dimensional (row or column)
- Aligning items in a single row or column
- Works along one axis (main axis & cross axis)
- Powerful for distributing space and aligning items flexibly
-Items are placed automatically in a row/column flow
-Great for flexible, content-driven layouts (like navigation bars)
-No dedicated gap property (use margins)

# **Main Differences: Grid**

- Two-dimensional (rows and columns)
- Creating complex grid-based layouts
- Works along both horizontal and vertical axes simultaneously
- Precise control over layout placement using grid lines
- Items can be explicitly positioned using grid lines and named areas
- More suited for structured layouts that need a specific arrangement
- Has grid-gap, row-gap, and column-gap for spacing



Q2.Explain the role of the following key properties in the Flexbox layout mode.
- justify-conten
- align-item
- gap
- flex-direction
- flex-wrap

justify-content :- Controls the horizontal alignment of flex items along the main axis

Common values:-
- flex-start (default) → Items align at the start.
- flex-end → Items align at the end.
- center → Items align in the center.
- space-between → Items spread out, first and last touching edges.
- space-around → Items have equal space around them.
- space-evenly → items have left right have equal space.

align-items :- Controls vertical alignment of flex items along the cross axis.

Common values:-
- flex-start → Aligns items at the top.
- flex-end → Aligns items at the bottom.
- center → Aligns items in the middle.
- stretch (default) → Stretches items to fill the container height.
- baseline → Aligns items based on their text baseline.

**gap** :- Defines space between flex items (without needing margins).

Shorthand for row-gap and column-gap.




```python
.container{
  diplay:flex;
  gap:20px;
}

```

**flex-direction:**

row (default) → Items placed left to right.
  - row-reverse → Items placed right to left.
  - column → Items placed top to bottom.
  - column-reverse → Items placed bottom to top.

  

```python
.container{
    display:flex;
    direction-flex:row;
}

```

 **flex-wrap** :- Controls whether items should wrap when they exceed the container width.

    - Common values:-
      - nowrap (default) → Items stay on one line (shrinking if needed).
      - wrap → Items move to the next line if they don't fit.
      - wrap-reverse → Similar to wrap, but wraps in reverse order.
